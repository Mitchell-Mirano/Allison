## A comparison between the different regressors

### Importing the necessary modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from allison.linear_models.linear_regression import LinearRegression
from allison.linear_models.polinomic_regression import PolinomicRegression
from allison.linear_models.functional_regression import FunctionalRegression
from allison.utils.functions.loss import mean_squared_error
from allison.utils.metrics import r2_score
```

### Linear Regression vs FunctionalRegression


```python
x=np.linspace(0,20,100)
y=x+ 2*np.sin(x) +0.5*np.random.randn(100)

plt.figure(figsize=(12,8))
plt.scatter(x,y)
```

```python
model=LinearRegression(loss_function=mean_squared_error,
                lr=0.001,
                metric=r2_score)

history=model.train(n_iters=30,
            features=x,
            labels=y,
            callbacks_period=5,history_train=True)
```

```python

plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.plot(x,model.predict(x),lw=4,color="red")
```

```python
model=FunctionalRegression()
model.optimizers(base_functions=["x","sinx"],
                loss_function=mean_squared_error,
                lr=0.001,
                metrics=r2_score)
                
history=model.train(n_iters=30,
                    features=x,
                    labels=y,
                    callbacks_period=5)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.plot(x,model.predict(x),lw=4,color="red")
```

### A polynomial of degree 1 vs a polynomial of degree 2

```python
x=np.linspace(0,10,100)
y=10-x**2 + 2*np.random.randn(100)
plt.figure(figsize=(12,8))
plt.scatter(x,y)
```

```python
model=PolinomicRegression(n_grade=1,
                loss_function=mean_squared_error,
                lr=0.001,
                metric=r2_score)
history=model.train(n_iters=100,
                    features=x,
                    labels=y,
                    callbacks_period=5,history_train=True)
```

```python

plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.plot(x,model.predict(x),lw=4,color="red")
```

```python
model=PolinomicRegression(n_grade=2,
                loss_function=mean_squared_error,
                lr=0.0001,
                metric=r2_score)

history=model.train(n_iters=200,
                    features=x,
                    labels=y,
                    history_train=True)
```

```python

plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.plot(x,model.predict(x),lw=4,color="red")
```
