### Importing the necessary modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from allison.regression import FunctionalRegression
from allison.utils.functions.loss import mean_squared_error
from allison.utils.metrics import r2_score
```

```python
x=np.linspace(0,20,1000)
y=x+ 2*np.sin(x) +0.5*np.random.randn(1000)

plt.figure(figsize=(12,8))
plt.scatter(x,y)
```

```python
model=FunctionalRegression(
                base_functions=["sinx","x"],
                loss_function=mean_squared_error,
                lr=0.0001,
                metric=r2_score
)

                
history=model.train(n_iters=500,
                    features=x,
                    labels=y,
                    callbacks_period=5,history_train=True)
```

```python
y_pred=model.predict(x)
r2 = model.metric(y, y_pred)
plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.plot(x,model.predict(x),lw=4,color="red")
plt.title(f'Functional Regression on Test Data(R2:{r2*100:.2f}%)')
```
